```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def func(x):
    return (x + 3)**2

def derivative(x):
    return 2 * (x + 3)

# Implement gradient descent
def gradient_descent(start_x, learning_rate, epochs):
    x = start_x
    xs = [x]
    ys = [func(x)]
    for i in range(epochs):
        grad = derivative(x)
        x -= learning_rate * grad
        xs.append(x)
        ys.append(func(x))
    return xs, ys

# Parameters for gradient descent
start_x = 2  # Starting x value
learning_rate = 0.1  # Learning rate
epochs = 20  # Number of gradient descent iterations

# Perform gradient descent
xs, ys = gradient_descent(start_x, learning_rate, epochs)

# Plotting the function and the gradient descent steps
x_values = np.linspace(-10, 10, 100)
y_values = func(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Function y = (x + 3)^2')
plt.scatter(xs, ys, color='red', label='Gradient Descent Steps')
plt.title('Gradient Descent Visualization')
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend()
plt.grid()
plt.show()
```
